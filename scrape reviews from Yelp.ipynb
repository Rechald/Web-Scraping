```python
# find all other paginations
def find_pages(myurl):
    # return all related paginations
    html = urllib.request.urlopen(myurl).read().decode('utf-8')
    pagination_marker = '<a class="available-number pagination-links_anchor" href="'
    pagination_links = [] 
    end = 0 
    while True:
        page_0 = html.find(pagination_marker, end)
        if page_0 == -1: break 
        page_start = page_0 + len(pagination_marker)
        page_end = html.find('">', page_start)
        pagination_links.append(html[page_start:page_end])
        end = page_end 

    return pagination_links
```

```python
def find_reviews(myurl):
    # show all reivews in the current url 
    html = urllib.request.urlopen(myurl).read().decode('utf-8')
    start = html.find('<script type="application/ld+json">')
    end = html.find('</script>', start)
    # find restaurant name 
    res_marker = '"name": "'
    start_0 = html.find(res_marker, start)
    res_start = start_0 + len(res_marker)
    res_end = html.find('"', res_start)
    res_name = html[res_start:res_end]
    # find all reviews
    results = ''
    cnt = 0 
    while True:
        content_mark = html.find('"description":', start)
        if content_mark == -1 or content_mark > end: 
            break 
        author_mark = html.find('"author":', content_mark)

        content_start = content_mark + len('"description": "')
        content_end = author_mark - len('", "')
        author_start = author_mark + len('"author": "')
        author_end = html.find('"', author_start)
        # new review
        start = author_end
        cnt += 1
        reviewer = html[author_start:author_end]
        content = html[content_start:content_end+1]
        print('reviewer ', cnt, ': ', html[author_start:author_end],  
              '\ncontent:', html[content_start:content_end+1],'\n')
        results += res_name + '\t' + reviewer + '\t' + content + '\n'
    return results
    
# Step 1: find all pagination links
other_pages = find_pages(yelp_restaurant_url)
all_pages = [yelp_restaurant_url] + other_pages
data_frame = 'restaurant_name \t reviewer \t content \n' 
for page in all_pages:
    data_frame += find_reviews(page)
# print(find_reviews(all_pages[0]))
output_fn = 'Assignment_1_Part_B.csv'

f = open(output_fn, 'w')
f.write(data_frame)
f.close()
print('Done')
```
